```python
from sklearn.metrics import accuracy_score, mean_absolute_error
from category_encoders import OrdinalEncoder
from sklearn.impute import SimpleImputer
from lightgbm import LGBMRegressor
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split, RandomizedSearchCV
```

```python
pipeline = make_pipeline(
    OrdinalEncoder(),
    LGBMRegressor(n_jobs=-1, n_estimators=1000, random_state=42, objective='mae')
)    
```

```python
target = 'price'
features = x.columns.drop([target, 'id', 'longitude', 'latitude'])
features
```

```python
train, test = train_test_split(x, random_state=42, test_size=.2)
```

```python
X_train = train[features]
X_test = test[features]
y_train = train[target]
y_test = test[target]
```

```python
pipeline.fit(X_train, y_train)
```

```python
base_predictions = x['price'].mean()
base_pred = [base_predictions] * len(y_test)
mean_absolute_error(base_pred, y_test)
```

```python
y_pred = pipeline.predict(X_test)
mean_absolute_error(y_test, y_pred)
```
